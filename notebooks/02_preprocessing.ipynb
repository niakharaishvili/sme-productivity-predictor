# 02 — Preprocessing & Feature Engineering
7 feature groups + 9 interaction terms

```python
import sys; sys.path.append('../src')
import pandas as pd
from data_loader import load_data, FEATURES, FEATURE_GROUPS, TARGET
from features import add_interactions, build_preprocessor, INTERACTION_COLS, ALL_FEATURES
from model import split
df = load_data()
print('Raw shape:', df.shape)
```

## 2.1 Feature Groups

```python
for group, cols in FEATURE_GROUPS.items():
    print(f'\n{group} ({len(cols)} vars):')
    for c in cols: print(f'  {c}')
```

## 2.2 Add Interaction Terms

```python
df = add_interactions(df)
print('Interaction features added:')
for c in INTERACTION_COLS: print(f'  {c}: mean={df[c].mean():.2f}')
```

## 2.3 Feature Matrix

```python
X = df[[c for c in ALL_FEATURES if c in df.columns]]
y = df[TARGET]
print('X shape:', X.shape)
print('\nClass distribution:')
print(y.value_counts())
```

## 2.4 Train/Test Split

```python
X_train, X_test, y_train, y_test = split(X, y)
print(f'Train: {X_train.shape} | Test: {X_test.shape}')
```

## 2.5 Preprocessor

```python
prep = build_preprocessor()
X_train_t = prep.fit_transform(X_train)
X_test_t  = prep.transform(X_test)
print('Transformed shape:', X_train_t.shape)
```
