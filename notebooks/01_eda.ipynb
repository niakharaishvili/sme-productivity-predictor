## Correlation Heatmap

```python
numeric_cols = ['firm_size','years_in_operation','digitization_score','innovation_index','training_investment']
fig, ax = plt.subplots(figsize=(8,6))
sns.heatmap(df[numeric_cols].corr(), annot=True, fmt='.2f', cmap='coolwarm', ax=ax)
plt.title('Feature Correlation Matrix')
plt.tight_layout()
plt.show()
```
